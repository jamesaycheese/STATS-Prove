```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
num_samples = 1000  # Number of samples to draw
num_trials = 1000   # Number of trials to perform
sample_size = 30    # Size of each sample

# Generate samples from a uniform distribution (could be any distribution)
samples = np.random.rand(num_trials, sample_size)

# Calculate the mean of each sample
sample_means = np.mean(samples, axis=1)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')

# Plot the theoretical normal distribution with the same mean and standard deviation
mu = np.mean(sample_means)
sigma = np.std(sample_means)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2)), color='b')

# Add labels and title
plt.title('Central Limit Theorem')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.show()
```

Explanation:

Parameters: We define the number of samples (num_samples), the number of trials (num_trials), and the size of each sample (sample_size).
Generating Samples: We generate num_trials samples, each containing sample_size random numbers drawn from a uniform distribution (although any distribution could be used).
Calculating Sample Means: We calculate the mean of each sample, resulting in an array of sample means.
Plotting: We plot a histogram of the sample means. The histogram represents the distribution of the sample means.
Theoretical Normal Distribution: We plot a theoretical normal distribution with the same mean and standard deviation as the sample means. According to the Central Limit Theorem, as the number of trials increases, the distribution of sample means approaches a normal distribution.
Visualization: The resulting plot demonstrates how the distribution of sample means becomes increasingly bell-shaped and approaches a normal distribution as the number of trials increases, regardless of the original distribution of the samples.
This example illustrates the Central Limit Theorem in action, showing how the distribution of sample means converges to a normal distribution with increasing sample size.
